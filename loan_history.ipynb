#### Top 10 Loans by Current Balance

```python
import plotly.express as px
import pandas as pd

# Grouping sample for demo
loan_sum = loan_data.groupby("Product Type")["Original Amt"].sum().reset_index()

# Bar chart
fig = px.bar(loan_sum, x="Product Type", y="Original Amt", title="Loan Amount by Product Type")

# Save as HTML
fig.write_html("loan_chart.html")

# Open it in your default browser
import webbrowser
webbrowser.open("loan_chart.html")
```

####  Active vs Paid Off Loans (Pie Chart)

```python
status_counts = loan_data["PaidOFF"].value_counts().rename({0: "Active", 1: "Paid Off"}).reset_index()
# Paid Off: This means the loan has been fully paid and is no longer active.  1=True
# Active: This means the loan is still active and has not been fully paid off yet. 0=False

status_counts.columns = ["Status", "Count"]
# Renames the columns to make them clearer. Instead of index and PaidOFF, we get "Status" and "Count"

fig = px.pie(status_counts, names="Status", values="Count", title="Loan Status: Active vs Paid Off", hole=0.4)

fig.write_html("loan_status_piechart.html")

import webbrowser
webbrowser.open("loan_status_piechart.html")
```

#### Payment Heatmap by Product Type and Payment Type

```python
import plotly.express as px

heatmap_data = loan_data.pivot_table(values='Pmt',
                                     index='Product Type',
                                     columns='PaymentType',
                                     aggfunc='mean').fillna(0)

fig = px.imshow(heatmap_data,
                labels=dict(x="Payment Type", y="Product Type", color="Avg Payment"),
                title="Average Payment by Product and Payment Type")

fig.write_html("payment_heatmap.html")

import webbrowser
webbrowser.open("payment_heatmap.html")
```

#### Maturity Status by Year

```python
# Visualize when loans are maturing (could show risk of bulk maturities).
# kitne loans mature ho rahe hain? jise dekh ke hum samajh sakte hain ki koi particular year mein zyada pressure to nahi hoga repayment ka.

loan_data['MaturityYear'] = loan_data['Maturity'].dt.year   # .dt.year = extracts only the year part
maturity_counts = loan_data['MaturityYear'].value_counts().sort_index().reset_index()

maturity_counts.columns = ['Year', 'LoanCount']

fig = px.bar(maturity_counts, x='Year', y='LoanCount',
             title="Loan Maturities by Year")

fig.write_html("maturities_by_year.html")

import webbrowser
webbrowser.open("maturities_by_year.html")
```
