```python
# Remove spaces and convert all columns to consistent lowercase
df.columns = df.columns.str.strip().str.lower()
print(df.columns.tolist())
```

#### Create Quarter coloumn AsofDate

```python
# Create quarter from date
df['asofdate'] = pd.to_datetime(df['asofdate'])  #convert date into datetime format
print(df['asofdate'].head())
# Create quarter column
df['quarter'] = df['asofdate'].dt.to_period('Q')   # Convert to quarter period for time series analysis
print(df['quarter'].head())
```

#### Calculate Risk matrics per quarter

```python
# Step 3: Group by correct column names
risk_summary = df.groupby('quarter').agg(
    total_loans=('acctno', 'nunique'),
    nonaccrual_loans=('nonaccrual', 'sum'),
    tdr_loans=('tdr', 'sum')
)
# count total loans in every quarter
# sum up the risky loans in every quarter
# sum up the TDR (risky) loans in every quarter
```

```python

# Step 4: Calculate percentages
risk_summary['% nonaccrual'] = (risk_summary['nonaccrual_loans'] / risk_summary['total_loans']) * 100
risk_summary['% tdr'] = (risk_summary['tdr_loans'] / risk_summary['total_loans']) * 100

# Step 5: Show result
print(risk_summary)
```

#### Total loans in 2022 Q2 is 1742
#### Non accural loans 2(unka interest record nhi ho rha)
#### tdr_loans	1 loan is TDR (Troubled Debt Restructured) 
#### % nonaccrual	Sirf 0.11% loans non-performing hain (bohat low risk)
#### % tdr	Sirf 0.05% loans restructured hain (again, low risk)

#### In 2020 Q2 --> the loan portfolio is very healthy — risk indicators are very low

```python
import matplotlib.pyplot as plt

risk_summary[['% nonaccrual', '% tdr']].plot(kind='line', marker='o')
plt.title("Risk Trends Over Time")
plt.xlabel("Quarter")
plt.ylabel("Percentage (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### Blue dot (% nonaccrual): 0.11% are non-performing loans
##### Orange dot (% TDR): 0.057% loans are restructured
##### both these values are from 2020 of 2nd quarter (Q2)
